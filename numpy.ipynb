## Question 4image.png

$$ P/ (A \sigma \epsilon T_0 ^ 4) = \left( \frac{1} {1+ e^{-kt}} \right)^4 $$

---



```python
kt = np.linspace(0,3,100)
#P to represent the dimenstionless quantity (as a whole) on the left side
P = (1/(1+np.exp(-kt)))**4
E = np.cumsum(P) * (kt[1]-kt[0]) #Finding the integral
```

```python
plt.plot(kt,E)
plt.xlabel('$kt$')
plt.ylabel(r'$\left(\frac{k}{A \sigma \epsilon T_0^4}\right) E(kt)$')
```
